```python
%%sql

DROP TABLE IF EXISTS raw_data.boxoffice;

CREATE TABLE raw_data.boxoffice (
  movieCd integer,
  showRange date,
  rank integer,
  salesAmt BIGINT,
  salesShare float,
  salesInten BIGINT,
  salesChange float,
  salesAcc BIGINT,
  audiCnt integer,
  audiInten integer,
  audiChange float,
  audiAcc BIGINT,
  scrnCnt integer,
  showCnt integer,
  PRIMARY KEY (showRange, rank)
);
```
